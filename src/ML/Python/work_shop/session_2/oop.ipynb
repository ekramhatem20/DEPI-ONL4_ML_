```python
class member:
    
    not_allowed_names = ["Hell","Shit","Baloot"]
    
    users_number = 0 
    @classmethod
    def show_count_users(cls):
        
        print(f"We have {cls.users_number} users in our system")
        
    
    @staticmethod
    def say_hello():
        
        print(f"hello uesr")
    
        
    def __init__(self,first_name,middle_name,last_name,gender):
        
         self.f_name = first_name
         
         self.m_name = middle_name
         
         self.L_name = last_name
         
         self.gender = gender
         
         member.users_number += 1
         
    def full_name(self):
        
        if self.f_name in member.not_allowed_names:
            
            raise ValueError("Name not allowed")
        
        return f"{self.f_name} {self.m_name} {self.L_name}"
    
    def name_with_title(self):
        
        if self.gender == "male":
            
            return f"Hello Mr {self.f_name}"
        elif self.gender == "female":
            
            return f"Hello Ms {self.f_name}"
        
        else:
            return f"Hello {self.f_name}"
        
    def get_all_info(self):
        
        return f"{self.name_with_title()} , Your full name: {self.full_name()}"
       
    def delet_user(self):
            
            member.users_number -= 1
            
            return f"User {self.f_name} Deleted"
```

```python
member_1 = member("ahmed","ali","kamel","male")
member_2 = member("alaa","ramadan","mohamed",'male')
member_3 = member("Malak","ahmed","sayed","female")
print(member_3.get_all_info())
print(member.users_number)
print(member_1.delet_user())
print(member.users_number)
member.show_count_users()
```

```python
class Member:
    
    def __init__(self,name):
        
        self.__name = name    #private
        
    def say_hello(self):
        
        return f"Hello {self.__name}"
    
    def get_name(self):
        
        return self.__name
    
    def set_name(self,new_name):
    
        self.__name = new_name
        
one = Member("ekram")
one.set_name('Abbas')

print(one.get_name())
print(one.say_hello())

        
```
