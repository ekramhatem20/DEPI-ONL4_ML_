```python
def factorial_1(num):
    if num == 0:
        return 1
    if num ==1:
        return 1 
    return  (num * factorial_1(num-1))
    
```

```python
factorial_1(8)
```

```python
def is_prime(num):
    
    for i in range(2,num):
        if num % i == 0 :
            return False
    return True

is_prime(6)
```

```python
def revers_text(txt):
    
    split_txt = txt.split(" ")
    new_list = split_txt[::-1]
    result = " ".join(new_list)
    return result
```

```python
revers_text("hatem ekram")
```

```python
import random
def Generate_password(length):
    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
    password = ""
    for _ in range (length):
        password += random.choice(characters)
    return f"Generate_password: {password} "
Generate_password(8)
```
